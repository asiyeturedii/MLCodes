```python
# This line is for enabling interactive plots in Jupyter notebook
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split

# Load the two CSV files from your desktop and display their first few rows
fruit1 = pd.read_csv('/Users/asiyerana/Desktop/fruit.csv')
fruit2 = pd.read_csv('/Users/asiyerana/Desktop/fruit2.csv')

# Display the first few rows of each dataset
fruit1.head()  # Shows the first 5 rows of the 'fruit.csv' file
fruit2.head()  # Shows the first 5 rows of the 'fruit2.csv' file

```

```python
lookup_fruit_name = dict(zip(fruit1.fruit_label.unique() , fruit1.fruit_name.unique()))
lookup_fruit_name
```

```python
# Create train-test split

X = fruit1[['mass','width','height']]
y= fruit1['fruit_label']

X_train , X_test , y_train , y_test = train_test_split(X, y, random_state=0)
```

```python
from matplotlib import cm
from pandas.plotting import scatter_matrix
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cmap = cm.get_cmap('gnuplot')

# Create scatter matrix
scatter = scatter_matrix(X_train, c=y_train, marker='o', hist_kwds={'bins': 15}, figsize=(12, 12), s=40, cmap=cmap)

# Show the plot
plt.show()
```
